```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.datasets import fetch_openml
import pandas as pd
boston = fetch_openml(name="boston", version=1, as_frame=True)
df = boston.frame
```

```python
x_min, x_max = df["LSTAT"].min(), df["LSTAT"].max()
x_1 = np.linspace(x_min, x_max, 100)
x_2 = np.zeros_like(x_1) + 1
X = np.vstack((x_1, x_2)).T
```

```python
plt.figure(figsize=(10, 6))
plt.scatter(df['LSTAT'], df['MEDV'])
# below is optional
plt.xlabel('% Lower Status Population (LSTAT)')
plt.ylabel('Median Value in $1000s (MEDV)')
plt.title('Housing Price vs. Lower Status Population')
```

```python
def neuron2(x,w):
    return x@w
```

```python
w2 = np.array([4.61, 1])  # Negative slope, intercept 30
Y2 = neuron2(X, w2)
```

```python
# Define our mean squared error loss function
def mse_loss(X, y, w):
    y_pred = neuron2(X, w)
    squared_errors = (y_pred - y) ** 2
    return np.mean(squared_errors)
```

```python
# Define a function to calculate the gradient of MSE loss
def mse_gradient(X, y, w):
    y_pred = neuron2(X, w)
    errors = y_pred - y
    gradient = (2/len(y)) * X.T @ errors
    return gradient
```
